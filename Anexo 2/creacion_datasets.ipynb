# Creaci칩n de los 3 datasets

```python
import os
import pandas as pd
```

```python
dirr_income = 'C:/Users/admin/Desktop/Correcciones/Income'
dirr_balance = 'C:/Users/admin/Desktop/Correcciones/Balance'
dirr_cashflow = 'C:/Users/admin/Desktop/Correcciones/Cashflow'

acciones_income = [nombre for nombre in os.listdir(dirr_income) if os.path.isfile(os.path.join(dirr_income, nombre))]
acciones_balance = [nombre for nombre in os.listdir(dirr_balance) if os.path.isfile(os.path.join(dirr_balance, nombre))]
acciones_cashflow = [nombre for nombre in os.listdir(dirr_cashflow) if os.path.isfile(os.path.join(dirr_cashflow, nombre))]

for accion in acciones_income:
    if accion.endswith(".ipynb"):
        acciones_income.remove(accion)

for accion in acciones_balance:
    if accion.endswith(".ipynb"):
        acciones_balance.remove(accion)

for accion in acciones_cashflow:
    if accion.endswith(".ipynb"):
        acciones_cashflow.remove(accion)
```

```python
columnas_income = ["Net Income", "Shares Outstanding (Basic)", "Dividend Per Share", "EBIT","EBITDA", "Revenue", "Accion"]
columnas_balance = ["Cash & Cash Equivalents", "Total Assets", "Total Liabilities","Shareholders' Equity","Total Long-Term Liabilities", "Total Current Liabilities","Long-Term Debt","Current Debt","Accion"]
columnas_cashflow = ["Net Cash Flow", "Free Cash Flow","Dividends Paid","Accion"]
```

```python
df_income = pd.DataFrame(columns=columnas_income)
df_balance = pd.DataFrame(columns=columnas_balance)
df_cashflow = pd.DataFrame(columns=columnas_cashflow)
```

```python
# Creaci칩n dataset income compilado

archivos_incompletos_income = []

for archivo in acciones_income:

    try:
        df_i = pd.read_excel(dirr_income + "/" + archivo)
        df_i = df_i.loc[:, df_i.columns != "Accion"]
        df_i = df_i.set_index("Quarter Ended").transpose()
        df_i = df_i[columnas_income[:-1]]
        df_i['Accion'] = archivo.split("-")[0]

        df_income = pd.concat([df_income, df_i], axis=0)
    
    except:
        archivos_incompletos_income.append(archivo)
```

```python
df_income['Accion'] = df_income['Accion'].apply(lambda x: x.upper())

df_income.head()
```

```python
# Verificaci칩n de variables faltantes en los archivos problem치ticos de Income

for archivo in archivos_incompletos_income:
    df_i_incompleto = pd.read_excel(dirr_income + "/" + archivo)
    df_i_incompleto = df_i_incompleto.loc[:, df_i_incompleto.columns != "Accion"]
    df_i_incompleto = df_i_incompleto.set_index("Quarter Ended").transpose()

    for columna in columnas_income[:-1]:
        if columna not in df_i_incompleto.columns:
            print(archivo.split("-")[0], columna)
```

```python
# Importar los dataset income, balance y cashflow para no ejecutar las celdas de arriba

import pandas as pd

df_income = pd.read_excel("df_income.xlsx")
df_balance = pd.read_excel("df_balance.xlsx")
df_cashflow = pd.read_excel("df_cashflow.xlsx")
```

```python
# Acciones comunes a los 3 datasets

conjunto_income = set(df_income['Accion'].unique())
conjunto_balance = set(df_balance['Accion'].unique())
conjunto_cashflow = set(df_cashflow['Accion'].unique())

acciones_comunes = conjunto_income.intersection(conjunto_balance).intersection(conjunto_cashflow)
acciones_comunes = sorted(list(acciones_comunes))

len(acciones_comunes)
```

```python
# Creaci칩n de la columna year para identificar a침os faltantes

df_income['Year'] = df_income['Date'].apply(lambda x: x.year) 
df_balance['Year'] = df_balance['Date'].apply(lambda x: x.year) 
df_cashflow['Year'] = df_cashflow['Date'].apply(lambda x: x.year) 
```

```python
years = ['Accion', 'First Year', 'Last Year', 'Missing Years', 'Total Years']

df_income_years = pd.DataFrame(columns=years)
df_balance_years = pd.DataFrame(columns=years)
df_cashflow_years = pd.DataFrame(columns=years)
```

```python
# Llenado df_income_years

for accion in acciones_comunes:
    
    anios = df_income.loc[df_income['Accion'] == accion, 'Year']
    anio_min = anios.min()
    anio_max = anios.max()
    diff_anios = anio_max - anio_min
    rango_anios = pd.Series(range(anio_min, anio_max + 1))
    anios_faltantes = rango_anios[~rango_anios.isin(anios)]

    df_income_years.loc[len(df_income_years.index)] = [accion, anio_min, anio_max, len(anios_faltantes), diff_anios]
```

```python
# Llenado df_balance_years

for accion in acciones_comunes:
    
    anios = df_balance.loc[df_balance['Accion'] == accion, 'Year']
    anio_min = anios.min()
    anio_max = anios.max()
    diff_anios = anio_max - anio_min
    rango_anios = pd.Series(range(anio_min, anio_max + 1))
    anios_faltantes = rango_anios[~rango_anios.isin(anios)]

    df_balance_years.loc[len(df_balance_years.index)] = [accion, anio_min, anio_max, len(anios_faltantes), diff_anios]
```

```python
# Llenado df_cashflow_years

for accion in acciones_comunes:
    
    anios = df_cashflow.loc[df_cashflow['Accion'] == accion, 'Year']
    anio_min = anios.min()
    anio_max = anios.max()
    diff_anios = anio_max - anio_min
    rango_anios = pd.Series(range(anio_min, anio_max + 1))
    anios_faltantes = rango_anios[~rango_anios.isin(anios)]

    df_cashflow_years.loc[len(df_cashflow_years.index)] = [accion, anio_min, anio_max, len(anios_faltantes), diff_anios]
```
